```python
#  Python code function for backward substitution
def get_backward_substitution(A):
    n = A.shape[0]
    x =np.zeros(n)

    for i in range(n - 1, -1, -1):
        x[i] = A[i, -1] / A[i, i]

        for j in range(i - 1, -1, -1):
            A[j, -1] -= A[j, i] * x[i]
            A[j, i] = 0

    return x
```

- Transform equations into an augmented matrix:
$$
\begin{align}

        1 & 0 & | & 3 \\
        1 & 1 & | & 1 \\
        0 & 1 & | & -2 \\
    \end{bmatrix}
\end{align}

- Transform matrix into its reduce row echelon form:
$$
\begin{align}
    \begin{bmatrix}

    \end{bmatrix}
    \xrightarrow{R_2 - R_1}
    \begin{bmatrix}

    \end{bmatrix} \\
    \begin{bmatrix}

        0 & 1 & | & -2 \\

    \end{bmatrix}
